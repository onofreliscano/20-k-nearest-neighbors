```python
# BLOQUE 01
# SETUP + IMPORTS + CARPETAS
# - Prepara librerías y rutas para un pipeline reproducible
# - Crea folders para guardar modelo, reportes y notas

import os
from pathlib import Path

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import matplotlib.pyplot as plt
import pickle

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

PROJECT_DIR = Path(".")
DATA_DIR = PROJECT_DIR / "data"
MODELS_DIR = PROJECT_DIR / "models"
REPORTS_DIR = PROJECT_DIR / "reports"

for d in [DATA_DIR, MODELS_DIR, REPORTS_DIR]:
    d.mkdir(parents=True, exist_ok=True)

print("OK dirs:", DATA_DIR, MODELS_DIR, REPORTS_DIR)

# Resultado esperado:
# Carpetas data/, models/, reports/ creadas
# Imports sin errores

# Interpretación:
# Base lista para cargar datos, entrenar y guardar artefactos
```

```python
# BLOQUE 02 (FIX)
# CARGA DEL DATASET (SEPARADOR CORRECTO)
# - El CSV está separado por ";" y venía “todo en una columna”
# - Se carga con sep=";" y se limpian comillas en headers

from pathlib import Path

NOTEBOOK_DIR = Path.cwd()
DATA_PATH = NOTEBOOK_DIR / "data" / "winequality-red.csv"

# comments: Fallback if executed from repo root or other cwd
if not DATA_PATH.exists():
    DATA_PATH = Path("src") / "data" / "winequality-red.csv"

if not DATA_PATH.exists():
    raise FileNotFoundError("Dataset not found in src/data/winequality-red.csv")

# comments: Read with correct separator and normalize column names
df = pd.read_csv(DATA_PATH, sep=";")
df.columns = [c.strip().replace('"', "") for c in df.columns]

print("Dataset path:", DATA_PATH.resolve())
print("Shape:", df.shape)
print("Columns:", list(df.columns))
df.head()

# Resultado esperado:
# Shape: (1599, 12)
# Columns: 11 features + 'quality'

# Interpretación:
# Dataset listo para crear el target (label) y separar X/y correctamente
```

```python
# BLOQUE 03 (FIX)
# CREACIÓN DEL TARGET 'label' + CHEQUEOS RÁPIDOS
# - El dataset trae 'quality' (score), no 'label' (0/1/2)
# - Convertimos quality -> label con bins y validamos distribución

# comments: Define target from 'quality' column
if "quality" not in df.columns:
    raise KeyError("Expected column 'quality' was not found in the dataset")

# comments: Map quality score to 3 classes (0/1/2) using bins
# 0 = Low (<=5), 1 = Medium (=6), 2 = High (>=7)
df["label"] = pd.cut(
    df["quality"],
    bins=[-np.inf, 5, 6, np.inf],
    labels=[0, 1, 2]
).astype(int)

target_col = "label"

print(df.info())
print("\nMissing values:\n", df.isna().sum().sort_values(ascending=False).head(20))
print("\nQuality (raw) counts:\n", df["quality"].value_counts().sort_index())
print("\nLabel (0/1/2) counts:\n", df[target_col].value_counts().sort_index())
print("\nLabel (0/1/2) %:\n", (df[target_col].value_counts(normalize=True).sort_index() * 100).round(2))

# Resultado esperado:
# Columna 'label' creada sin errores
# Conteos para label 0/1/2 visibles (posible desbalance)

# Interpretación:
# Ya tenemos el target correcto para KNN multiclase y sabemos si hay clases minoritarias
```

```python
# BLOQUE 04
# SEPARAR X / y
# - Define features (X) sin el target
# - Define y como la columna label

X = df.drop(columns=[target_col]).copy()
y = df[target_col].copy()

print("X shape:", X.shape)
print("y shape:", y.shape)
print("y unique:", sorted(y.unique()))
print("Features:", list(X.columns))

# Resultado esperado:
# X con 11 columnas y y con 1 columna/serie
# y con valores [0, 1, 2]

# Interpretación:
# Ya tenemos inputs y target listos para split y escalado
```

```python
# BLOQUE 05
# TRAIN/TEST SPLIT 80/20 (STRATIFY)
# - Separa datos para medir generalización
# - Mantiene proporción de clases con stratify

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

print("Train:", X_train.shape, y_train.shape)
print("Test :", X_test.shape, y_test.shape)
print("\nTrain %:\n", (y_train.value_counts(normalize=True).sort_index() * 100).round(2))
print("\nTest  %:\n", (y_test.value_counts(normalize=True).sort_index() * 100).round(2))

# Resultado esperado:
# Shapes correctos y proporciones similares train vs test

# Interpretación:
# Evitamos un split sesgado que infle o hunda el accuracy
```

```python
# BLOQUE 06
# ESCALADO (STANDARD SCALER)
# - KNN usa distancias: escalado es crítico
# - Fit SOLO en train (evita leakage) y transform en test

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaled train:", X_train_scaled.shape)
print("Scaled test :", X_test_scaled.shape)

# Resultado esperado:
# Matrices escaladas con mismas dimensiones de X_train/X_test

# Interpretación:
# Todas las variables aportan de forma comparable al cálculo de distancia
```

```python
# BLOQUE 07
# KNN BASELINE (k=5)
# - Entrena un modelo inicial para tener línea base
# - Sirve para comparar contra la optimización de k

k_initial = 5
knn_baseline = KNeighborsClassifier(n_neighbors=k_initial)

knn_baseline.fit(X_train_scaled, y_train)
y_pred_baseline = knn_baseline.predict(X_test_scaled)

acc_baseline = accuracy_score(y_test, y_pred_baseline)
acc_baseline

# Resultado esperado:
# Un accuracy baseline (varía por dataset)

# Interpretación:
# Tenemos un piso de performance antes del tuning
```

```python
# BLOQUE 08
# EVALUACIÓN BASELINE
# - Mide accuracy, confusion matrix y classification report
# - Identifica clases difíciles (0 vs 1 vs 2)

cm_baseline = confusion_matrix(y_test, y_pred_baseline)
report_baseline = classification_report(y_test, y_pred_baseline, digits=3)

print("Baseline k =", k_initial)
print("Accuracy:", round(acc_baseline, 4))
print("\nConfusion Matrix:\n", cm_baseline)
print("\nClassification Report:\n", report_baseline)

# Resultado esperado:
# Matriz 3x3 y reporte con precision/recall/f1 por clase

# Interpretación:
# Vemos si el modelo confunde especialmente baja vs media o media vs alta
```

```python
# BLOQUE 09
# OPTIMIZAR k (1..20)
# - Prueba distintos k y guarda accuracy en lista
# - Selecciona best_k por máximo accuracy en test

k_values = list(range(1, 21))
acc_by_k = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_scaled, y_train)
    preds = model.predict(X_test_scaled)
    acc_by_k.append(accuracy_score(y_test, preds))

best_idx = int(np.argmax(acc_by_k))
best_k = k_values[best_idx]
best_acc = acc_by_k[best_idx]

results = pd.DataFrame({"k": k_values, "accuracy": acc_by_k}).sort_values("accuracy", ascending=False)
print("Best k:", best_k, "| Best accuracy:", round(best_acc, 4))
results.head(10)

# Resultado esperado:
# Tabla con los mejores k en top 10
# best_k y best_acc impresos

# Interpretación:
# Elegimos k por evidencia y no por intuición
```

```python
# BLOQUE 10
# PLOT ACCURACY vs k
# - Visualiza trade-off: k pequeño (más ruido) vs k grande (más suavizado)
# - Facilita escoger k óptimo de forma clara

plt.figure(figsize=(10, 5))
plt.plot(k_values, acc_by_k, marker="o")
plt.xticks(k_values)
plt.xlabel("k (n_neighbors)")
plt.ylabel("Accuracy")
plt.title("KNN Wine Quality: Accuracy vs k")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# Resultado esperado:
# Curva accuracy vs k visible y legible

# Interpretación:
# Se observa dónde el modelo se estabiliza o cae por exceso de suavizado
```

```python
# BLOQUE 11
# MODELO FINAL (best_k) + EVALUACIÓN FINAL
# - Re-entrena con best_k para el resultado final
# - Reporta métricas finales para entrega

knn_final = KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(X_train_scaled, y_train)

y_pred_final = knn_final.predict(X_test_scaled)

acc_final = accuracy_score(y_test, y_pred_final)
cm_final = confusion_matrix(y_test, y_pred_final)
report_final = classification_report(y_test, y_pred_final, digits=3)

print("Final k =", best_k)
print("Accuracy:", round(acc_final, 4))
print("\nConfusion Matrix:\n", cm_final)
print("\nClassification Report:\n", report_final)

# Resultado esperado:
# Accuracy final y reporte por clase del modelo óptimo

# Interpretación:
# Este es el performance que vas a comunicar y guardar como modelo final
```

```python
# BLOQUE 12
# GUARDAR MODELO + SCALER + RESULTADOS DE TUNING
# - Guarda artefactos para inferencia reproducible
# - Deja evidencia (CSV) del tuning k vs accuracy

model_path = MODELS_DIR / f"knn_wine_classifier_k{best_k}.pkl"
scaler_path = MODELS_DIR / "standard_scaler.pkl"
tuning_csv_path = REPORTS_DIR / "knn_accuracy_by_k.csv"

with open(model_path, "wb") as f:
    pickle.dump(knn_final, f)

with open(scaler_path, "wb") as f:
    pickle.dump(scaler, f)

pd.DataFrame({"k": k_values, "accuracy": acc_by_k}).to_csv(tuning_csv_path, index=False)

print("Saved model ->", model_path)
print("Saved scaler ->", scaler_path)
print("Saved tuning ->", tuning_csv_path)

# Resultado esperado:
# 2 archivos .pkl en models/
# 1 CSV de tuning en reports/

# Interpretación:
# El modelo queda listo para usarse sin reentrenar y con trazabilidad QA
```

```python
# BLOQUE 06B (FIX DEFINITIVO)
# RE-ESCALADO Y RE-ENTRENAMIENTO SIN 'quality'
# - El scaler y el modelo se ajustan solo con las 11 features
# - Evita errores de feature names en inferencia

FEATURES = df.drop(columns=["quality", "label"]).columns.tolist()

X = df[FEATURES]
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_final = KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(X_train_scaled, y_train)

# Resultado esperado:
# Scaler y modelo entrenados solo con 11 features
# Sin warnings ni errores de nombres

# Interpretación:
# Pipeline limpio y consistente entre entrenamiento e inferencia
```
